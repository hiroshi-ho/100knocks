```python
import os
import re
fname = "nlp.txt"

def nlp_lines():
    with open(fname) as lines:
        pattern = re.compile(r"(^.*?[\.|;|:|\?|!])\s([A-Z].*)")
        for line in lines:
            line = line.strip()
            while len(line) > 0:
                #行から一文を取得
                match = pattern.match(line)
                if match:
                    yield match.group(1)
                    line = match.group(2)
                else:
                    yield line
                    line = ""
            #print(line)
for line in nlp_lines():
    print(line)
```
